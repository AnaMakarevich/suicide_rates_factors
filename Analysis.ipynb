```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Basic Exploratory Analysis

```python
df = pd.read_csv('full_suicide_dataset.csv')
```

```python
cols = list(set(df.columns)-{'country', 'suicide_rate'})
just_values = df[cols]
```

```python
for i in range(len(just_values.columns)):
    print("Plot for: " + just_values.columns[i])
    corr = np.corrcoef(just_values.iloc[:,i], df.suicide_rate)
    print("Correlation coefficient: " + str(corr[0,1]))
    plt.scatter(just_values.iloc[:,i], df.suicide_rate)
    plt.xlabel(just_values.columns[i])
    plt.ylabel('suicide rate')
    plt.rcParams["figure.figsize"] = (8,8)
    plt.show()
```

```python
import seaborn as sns
corrdf = df.corr()
sns.heatmap(corrdf)
```
